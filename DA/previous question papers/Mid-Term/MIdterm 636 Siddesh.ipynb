```r
airports<-na.omit(airports)
```

```r
city_airports<-airports %>% group_by(city) %>% tally()

city_airports
```

```r
barplot(city_airports$n, main="City Airports",
   xlab="Name of Airports",ylab = "Number of Airports",
names.arg =city_airports$city)
```

```r
head(airports)
```

```r
airports <- transform(airports, dst = ifelse(dst == "A", "USA",dst))
airports <- transform(airports, dst = ifelse(dst != "USA", "Others",dst))
```

```r
head(airports)
```

```r
city_usa_airports<-airports %>% group_by(city,dst) %>% tally()
colnames(city_usa_airports)[2] <- "USA"
```

```r
city_usa_airports
```

```r
barplot(city_usa_airports$n, main="City Airports USA",
   xlab="Name of Airports",ylab = "Number of Airports",names.arg =city_usa_airports$city)
```
