````r
#Load the data into dataframe and retrieve the "Adj Close" column values
df.ibm = data.frame(read.csv("IBM.csv"))[,6]
df.aapl = data.frame(read.csv("AAPL.csv"))[,6]
df.amzn = data.frame(read.csv("AMZN.csv"))[,6]
df.fb = data.frame(read.csv("FB.csv"))[,6]
df.goog = data.frame(read.csv("GOOG.csv"))[,6]
df.msft = data.frame(read.csv("MSFT.csv"))[,6]
````

````r
#user built correlation function using given formula
corfun = function(x,y){
    n = sum((x - mean(x)) * (y - mean(y)))
    d = sqrt(sum((x - mean(x)) ** 2) * sum((y - mean(y)) ** 2))
    r = n / d
    print(r)
}
````

````r
all=list(df.ibm,df.msft,df.aapl,df.amzn,df.fb,df.goog) #creating a list of all companies
cor.ibm = rbind(sapply(all,corfun,all[[1]]))
cor.msft = rbind(sapply(all,corfun,all[[2]]))
cor.aapl = rbind(sapply(all,corfun,all[[3]]))
cor.amzn = rbind(sapply(all,corfun,all[[4]]))
cor.fb = rbind(sapply(all,corfun,all[[5]]))
cor.goog = rbind(sapply(all,corfun,all[[6]]))
cor.all = rbind(cor.ibm,cor.msft,cor.aapl,cor.amzn,cor.fb,cor.goog)
names = c('IBM', 'MSFT', 'AAPL', 'AMZN', 'FB', 'GOOG')
rownames(cor.all) = names
colnames(cor.all) = names
cor.all #matrix of all corellations between 2 companies
````

````r
count = 1
result = matrix(nrow = 15, ncol = 2)
for(i in 2:6){
    for(j in 1:(i-1)){
        result[count,1] = paste(names[j] ,':', names[i],'=')
        result[count,2] = round(cor.all[i,j],3)
        count = count + 1
    }
}
````
