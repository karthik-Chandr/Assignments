```r
#1.22 
#2. primes[-n] return all the elements in the data set "prime" except the last element.
n <- length(primes)
primes[-n]
```

```r
#1.22
#3. To find the number of twin primes, primes[-1]-primes[-n] gives us a vector with the difference between two neighbours 
#   which is 2.if we find the number of elements equal to 2 then we find the number of twin primes
sum.(primes[-1]-primes[-n]==2)
```
